```python
# Exponential distribution inference
import matplotlib.pyplot as plt
import random

N = 20000000

p = 0.0005 
# 0.9995
threshold = 1 * (1.0 - p)

def isSuccess(x):
    return x >= threshold


lastSuccess = 0
delay_time_arr = [] # 1, 4, 

for epoch in range(N):
    x = random.random() # 0~ 1
    if isSuccess(x):
        delay_time_arr.append(epoch-lastSuccess)
        lastSuccess = epoch

plt.hist(delay_time_arr,bins=1000)
plt.show()
    
```
