```python
# Poisson distribution inference
import matplotlib.pyplot as plt
import random
import math
import numpy as np
```

```python
TN = 20000000

N = 10000

p = 0.005 

threshold = 1 * (1.0 - p)

lamb = N * p

x = np.arange(0,100,1)
y = [ math.pow(lamb,x_) * math.exp(-lamb) / math.factorial(x_) for x_ in x]


```

```python
def isSuccess(x):
    return x >= threshold


success_count = 0
success_count_arr = []

for epoch in range(TN):
    rn = random.random()
    if isSuccess(rn):
        success_count += 1
    
    if epoch % N == 0 and epoch != 0:
        success_count_arr.append(success_count)
        success_count = 0

plt.hist(success_count_arr,bins=100)
plt.show()
```
